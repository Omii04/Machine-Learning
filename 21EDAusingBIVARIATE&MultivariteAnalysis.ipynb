```python
import pandas as pd          
import seaborn as sbn  
import matplotlib.pyplot as plt         
```

```python
titanic=pd.read_csv(r"C:\Users\om420\Downloads\titanic_sample_campusx.csv")
titanic
```

```python
#scatterplot (Numerical-Numerical)

sbn.scatterplot(x=titanic['Pclass'],y=titanic['Age'],hue=titanic['Fare'])
```

```python
#Barplot (Numerical-Categorical)
sbn.barplot(x=titanic['Pclass'],y=titanic['Fare'],hue=titanic['Sex'])
```

```python
#Boxplot (Numerical-categorical)

sbn.boxplot(x=titanic['Pclass'],y=titanic['Age'])
```

```python
# Distplot (Numerical-Categorical)
sbn.distplot(titanic['Survived']==0,hist=False)
sbn.distplot(titanic['Survived']==1,hist=False)
```

```python
# Heatmap (categorical-categorical)
sbn.heatmap(pd.crosstab(titanic['Sex'],titanic['Survived']))
(titanic.groupby('Sex').mean('Survived')*100)
```
