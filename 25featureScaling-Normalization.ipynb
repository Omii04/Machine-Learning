```python
import pandas as pd      
import numpy as np             
import matplotlib.pyplot as plt                      
import seaborn as sbn            
```

```python
df=pd.read_csv(r"C:\Users\om420\Downloads\student_scores_large.csv",usecols=[0,1,2])
df.columns=['M','E','S']
df
```

```python
sbn.kdeplot(df['E'])
```

```python
df=df.drop(['M'],axis=1)
```

```python
from sklearn.model_selection import train_test_split
X = df[['E']]      # double brackets to keep it 2D
y = df['S']

# Step 2: Split into train and test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler.fit(df)

scaled_x_train = scaler.fit_transform(X_train)
scaled_x_test = scaler.fit_transform(X_test)
```

```python
scaled_x_train = pd.DataFrame(scaled_x_train ,columns=X_train.columns, index=X_train.index)
X_test_scaled = pd.DataFrame(scaled_x_test, columns=X_test.columns, index=X_test.index)
```

```python
sbn.distplot([scaled_x_train],hist=False)
sbn.distplot([X_test_scaled],hist=False)
```
