```python
import pandas as pd
import numpy as np           
import seaborn as sbn             
import matplotlib.pyplot as plt

```

```python
df=pd.read_csv(r"C:\Users\om420\Downloads\data_science_job.csv")
df.head()
```

```python
df.isnull().mean()*100
```

```python

cols=[var for var in df.columns if df[var].isnull().mean()<0.05 and df[var].isnull().mean()>0]
cols
```

```python
len(df[cols].dropna())/len(df)      # Percentage of rows with missing values in selected columns
```

```python
new_df=df[cols].dropna()  # Create a new DataFrame with rows that have no missing values in the selected columns
```

```python
new_df.hist(bins=50,density=True,figsize=(12,12))
plt.show()
```

```python
df['training_hours'].hist(bins=50,color='red',density=True)   #before cca 
new_df['training_hours'].hist(bins=50,color='Blue',density=True)   #After cca
```

```python
df['experience'].hist(bins=50,color='red',density=True)  #before cca
new_df['experience'].hist(bins=50,color='Blue',density=True)   #after cca
```
