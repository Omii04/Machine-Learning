```python
import numpy as np                   
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df=pd.read_csv(r"C:\Users\om420\Downloads\placement.csv")
df.head()
```

```python
plt.figure(figsize=(16,5))
plt.subplot(1,2,1)
sns.distplot(df['cgpa'])

plt.subplot(1,2,2)
sns.distplot(df['placement_exam_marks'])

plt.show()
```

```python
df["placement_exam_marks"].describe()
```

```python
sns.boxplot(df["placement_exam_marks"])
```

```python
#finding the IQR
percentile25=df['placement_exam_marks'].quantile(0.25)

percentile75=df['placement_exam_marks'].quantile(0.75)
```

```python
iqr=percentile75-percentile25
iqr
```

```python
upper_limit=percentile75+1.5*iqr
lower_limit=percentile25-1.5*iqr
```

FINDING OUTLIERS


```python
df[df['placement_exam_marks']>upper_limit]
```

```python
df[df['placement_exam_marks']<lower_limit]
```

TRIMMING

```python
new_df=df[df['placement_exam_marks']<upper_limit]
new_df.shape
```

```python
plt.figure(figsize=(16,7))
plt.subplot(2,2,1)
sns.distplot(df['placement_exam_marks'])

plt.figure(figsize=(16,7))
plt.subplot(2,2,2)
sns.distplot(new_df['placement_exam_marks'])

plt.figure(figsize=(16,7))
plt.subplot(2,2,3)
sns.boxplot(df['placement_exam_marks'])

plt.figure(figsize=(16,7))
plt.subplot(2,2,4)
sns.boxplot(new_df['placement_exam_marks'])

plt.show()
```

CAPPING

```python
new_df_cap=df.copy()
```

```python
new_df_cap['placement_exam_marks']=np.where(new_df_cap['placement_exam_marks']>upper_limit,upper_limit,
                                            np.where(new_df_cap['placement_exam_marks']<lower_limit,lower_limit,
                                            new_df_cap['placement_exam_marks']))
```
