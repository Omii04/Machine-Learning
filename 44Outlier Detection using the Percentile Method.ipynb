```python
upper_limit = df.quantile(0.98)
upper_limit
```

```python
lower_limit = df.quantile(0.02)
lower_limit
```

```python
new_df=df[(df <= upper_limit) & (df >= lower_limit)]
new_df
```

```python
new_df.describe()
```

```python
sns.boxplot(new_df)
```

```python
sns.distplot(new_df)
```

```python
#CAPING- WINSORIZATION
dff=np.where(df>=upper_limit,upper_limit,
            np.where(df<=lower_limit,lower_limit,
                     df))
```

```python
sns.distplot(dff)
```
