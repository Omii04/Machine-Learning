```python
from sklearn.datasets import make_regression
import numpy as np        
import matplotlib.pyplot as plt          
```

```python
X,y=make_regression(n_samples=4,n_features=1,n_informative=1,n_targets=1,noise=80,random_state=42)
```

```python
plt.scatter(X,y)
```

```python
#lets apply ols
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
model.fit(X,y)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
plt.scatter(X,y)
plt.plot(X,model.predict(X),color='red')
```

```python
#lets apply gradient descent assuming slope is constant m=31.69
# and starting value of intercept b=0

y_pred = ((-57.99*X)+0).reshape(4)
```

```python
plt.scatter(X,y)
plt.plot(X,model.predict(X),color='red',label='ols')
plt.plot(X,y_pred,color="blue",label='b=0')
plt.legend()
plt.show()
```

```python
#iteration 1
m=78.54
b=0

loss_slope= -2*np.sum(y-m*X.ravel()-b)
loss_slope
```

```python
# lest take learning rate 0.1
lr=0.1

step_size=lr*loss_slope
step_size
```

```python
b=b-step_size
b
```

```python
y_pred1=((-57.99*X)+b).reshape(4)
plt.scatter(X,y)
plt.plot(X,model.predict(X),color='red',label='OLS')
plt.plot(X,y_pred1,color='black',label='b={}'.format(b))
plt.plot(X,y_pred,color='orange',label='b=0')
plt.legend()
plt.show()
```

```python
# Iteration 2
loss_slope= -2 * np.sum(y-m*X.ravel()-b)
loss_slope
```

```python
step_size=lr*loss_slope
step_size
```

```python
b=b-step_size
b
```

```python
y_pred2=((-57.99*X)+b).reshape(4)

plt.scatter(X,y)
plt.plot(X,model.predict(X),color='red',label='OLS')
plt.plot(X,y_pred1,color='grey',label='b={}'.format(b))
plt.plot(X,y_pred,color='grey',label='b=0')
plt.plot(X,y_pred2,color='black',label='b={}'.format(b))
plt.legend()
plt.show()
```

```python
b=100
m=-57.99
lr=0.1

epochs=100
for i in range(epochs):
    loss_slope=np.sum(-2*(y - m*X.ravel() - b))
    step_size=lr*loss_slope
    b=b-step_size
    
    y_pred=m*X+b
    
    plt.plot(X,y_pred)
    plt.scatter(X,y)
```
