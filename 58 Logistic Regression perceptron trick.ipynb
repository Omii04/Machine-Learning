```python
from sklearn.datasets import make_classification
import matplotlib.pyplot as plt
import numpy as np 
X, y = make_classification(
    n_samples=100,
    n_features=2,
    n_informative=2,
    n_redundant=0,
    random_state=41,
    n_classes=2,
    n_clusters_per_class=1
)
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
```
