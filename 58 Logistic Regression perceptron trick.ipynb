```python
from sklearn.datasets import make_classification
import matplotlib.pyplot as plt
import numpy as np 
X, y = make_classification(
    n_samples=100,
    n_features=2,
    n_informative=2,
    n_redundant=0,
    random_state=41,
    n_classes=2,
    n_clusters_per_class=1
)
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
```

```python
def perceptron(X,y):
    X=np.insert(X,0,1,axis=1)
    weight=np.ones(X.shape[1])
    lr= 0.1 
    
    for i in range(1000):
        j=np.random.randint(0,100)
        y_hat= step(np.dot(X[j],weight))
        weight=weight+lr*(y[j]-y_hat)*X[j]
        
    return weight[0],weight[1:]
```

```python
def step(z):
    return 1 if z>0 else 0
```

```python
intercept_,coef_=perceptron(X,y)
```

```python
print(intercept_)
print(coef_)
```

```python
m=(coef_[0]/coef_[1])
b=(intercept_/coef_[1])
```

```python
X_input=np.linspace(-3,3,100)
y_input=m*X_input+b
```

```python
plt.figure(figsize=(10,6))
plt.plot(X_input,y_input,color='red',linewidth=3)
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
plt.ylim(-3,2)
```

Using sklearn logistic regression

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
model.fit(X,y)
```

```python
m=-(model.coef_[0][0]/model.coef_[0][1])
b=-(model.intercept_/model.coef_[0][1])
```

```python
X_input1=np.linspace(-3,3,100)
y_input1=m*X_input1+b 
```

```python
plt.figure(figsize=(10,6))
plt.plot(X_input1,y_input1,color='green',linewidth=3)
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
plt.ylim(-3,2)
```

```python
plt.figure(figsize=(10,6))

# Plot both lines
plt.plot(X_input, y_input, color='red', linewidth=3, label='perceptron model')
plt.plot(X_input1, y_input1, color='green', linewidth=3, label='sklearn logistic Model')

# Plot the same scatter data
plt.scatter(X[:,0], X[:,1], c=y, cmap='winter', s=100)

plt.ylim(-3, 2)
plt.legend()
plt.title("Direct Boundary Comparison")
plt.show()
```
