```python
import numpy as np 
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt
```

```python
X,y=make_classification(n_samples=100,n_features=2,n_classes=2,n_informative=1,n_redundant=0,n_clusters_per_class=1,random_state=42,hypercube=False,class_sep=20)
```

```python
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter')
plt.show()
```

By Using Sklearn Logistic Regression

```python
lor=LogisticRegression(penalty=None,solver='saga')
lor.fit(X,y)
```

```python
print(lor.coef_)
print(lor.intercept_)
```

```python
m1=-(lor.coef_[0][0]/lor.coef_[0][1])
b1=-(lor.intercept_/lor.coef_[0][1])
```

```python
X_input1=np.linspace(-3,3,100)
y_input1=m1*X_input1+b1
```

```python
def gd(X,y):
    X=np.insert(X,0,1,axis=1)
    weight=np.ones(X.shape[1])
    lr=0.5
    
    for i in range(5000):
        y_hat=sigmoid(np.dot(X,weight))
        weight=weight+lr*(np.dot((y-y_hat),X))/X.shape[0]
        
    return weight[1:],weight[0]
        
```

```python
def sigmoid(z):
    return 1/(1+np.exp(-z))
```

```python
coef_,intercept_= gd(X,y)
```

```python
m=-(coef_[0]/coef_[1])
b=-(intercept_/coef_[1])
```

```python
X_input=np.linspace(-3,3,100)
y_input=m*X_input+b
```

```python
plt.figure(figsize=(10,6))
plt.plot(X_input1,y_input1,color='black',linewidth=3)
plt.plot(X_input,y_input,color='red',linewidth=3)
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
plt.show()
```
