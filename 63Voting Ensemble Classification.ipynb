```python
import numpy as np                  
import pandas as pd
```

```python
df=pd.read_csv(r"C:\Users\om420\Downloads\archive (4)\iris_synthetic_data.csv")
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
df['label']=LabelEncoder().fit_transform(df['label'])
df=df.tail(-1000)
```

```python
X=df.iloc[:,0:4]
y=df.iloc[:,-1]
y.unique()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import VotingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestClassifier
```

```python
clf1=LogisticRegression()
clf3=KNeighborsClassifier()
clf4=RandomForestClassifier()
```

```python
estimators=[('lr',clf1),('knn',clf3),('rf',clf4)]
```

```python
for estimator in estimators:
    scores = cross_val_score(estimator[1], X, y, cv=5, scoring='accuracy')
    print(estimator[0], np.round(scores.mean(), 3))
```

```python
vc=VotingClassifier(estimators=estimators)
X=cross_val_score(vc,X,y,cv=5,scoring='accuracy')
print(np.round(np.mean(X),2))
```
